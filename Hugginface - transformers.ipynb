## Hugging face Text translation

```python
translator = pipeline('translation_en_to_fr', model='google-t5/t5-base')

translation = translator("programming is fun!")[0]['translation_text']
print(translation)
```

```python
translation = translator("I love you!")[0]['translation_text']
print(translation)
```

Lets translate from dutch to english 

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-nl-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-nl-en")

text = 'Hallo vrienden, hoe gaat het vandaag?'
tokenized_text = tokenizer([text], return_tensors='pt')

translation = model.generate(**tokenized_text)
translated_text = tokenizer.batch_decode(translation, skip_special_tokens=True)[0]

print(translated_text)
```
